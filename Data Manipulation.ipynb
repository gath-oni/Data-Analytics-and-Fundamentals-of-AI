### **3. GROUP BY AND SUMMARIZE**

```python
print("="*100)
print(" GROUP BY AND SUMMARIZING")
print("="*100)

print("Mean Income by Education levels")

income_by_educ = df.groupby('degree')['income'].mean().sort_values(ascending = False)
income_by_educ
```

```python
print("="*100)
print(" GROUP BY AND SUMMARIZING")
print("="*100)

print("Median Income by Education levels")

income_by_educ = df.groupby('degree')['income'].median().sort_values(ascending = False)
income_by_educ
```

```python
print("="*100)
print(" GROUP BY AND SUMMARIZING")
print("="*100)

print("Mean Income by Marital Status")

income_by_mar = df.groupby('marital')['income'].mean().sort_values(ascending = False)
income_by_mar
```

```python
## Multiple grouping by and summarizing
print("="*100)
print(" GROUP BY AND SUMMARIZING")
print("="*100)

print("Mean Income by Education and Gender levels")

income_by_educ = df.groupby(['sex','degree'])['income'].mean()
income_by_educ
```

##### **Reset the index for the series**

```python
income_by_educ = df.groupby(['sex','degree'])['income'].mean().reset_index()
income_by_educ
```

##### **Unstack the table**

```python
income_by_educ = df.groupby(['sex','degree'])['income'].mean().unstack()
income_by_educ
```

```python
income_by_educ = df.groupby(['sex','marital','degree'])['income'].mean().unstack()
income_by_educ
```

##### **Advanced Grouping and Multiple Aggregation**

```python
summary = df.groupby('degree').agg({
    'income':['mean', 'median', 'std', 'var', 'count'],
    'age':['mean', 'median'],
    'height': ['mean', 'median']
})
summary
```

#### **FEATURE ENGINEERING**

```python
df.head()
```

##### **Create Income Categories**

```python
mean_income = df['income'].mean()
print(f"Mean: {mean_income}")
```

```python
df['Income_Category'] = np.where(df['income'] >= mean_income,
                                 'HIGH INCOME','LOW INCOME')
df.head()
```

##### **Frequency of Income Category**

```python
freq = df['Income_Category'].value_counts()
freq
```

##### **Descriptive Statistics**

```python
df.drop(columns = ['id']).describe()
```

```python
bins = [19, 35, 45, 55, float('inf')]
labels = ['YOUNG', 'YOUNG ADULT', 'MIDDLE ADULT', 'OLDER ADULT']
df['Age_Cat'] = pd.cut(df['age'],
                       bins = bins,
                       labels = labels,
                       right = False)
df.head()
```

```python
freq = df['Age_Cat'].value_counts()
freq
```
