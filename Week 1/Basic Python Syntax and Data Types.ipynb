#### **Mathematical Operations**

```python
a, b = 20, 9
print('='*60)
print("ARITHMETIC OPERATIONS")
print('='*60)
print(f"Add: {a+b} | Sub: {a-b} | Mul: {a*b}")
```

```python
print('='*60)
print("ADVANCED OPERATIONS")
print('='*60)
print(f"Div: {a/b} | Floor: {a//b} | Mod: {a%b} | Power: {a**2}")
```

#### **Logical Operators**

```python
score = 62
score
```

```python
## Conditional logic
if score >= 70:
    grade = "A"
elif score >= 60:
    grade = "B"
elif score >= 50:
    grade = "C"
elif score >= 40:
    grade = "D"
else:
    grade = "Fail"
```

```python
print(f"Score: {score}, Grade: {grade}")
```
